# Bibliotecas Python

### Formas de Importação:

1) Função específica:

```
from nome_biblioteca import nome_funcao
```



2. Importar a biblioteca toda:

```
import nome_biblioteca
```

```
from nome_biblioteca import *
```





3. Importar biblioteca com alias:

```
import nome_biblioteca as nome_alias
```

### Matplotlib

```
import matplotlib.pyplot as plt
```
- Importa a biblioteca matplotlib;
- Usa ``plt` para chamar as funções e métodos;
- Serve para plotar gráficos;

```
plt.bar(x= , height= )
```
- Plota gráfico de barras;
- Pode colocar apenas os dados como parametros (sem passar x e height)


```
plt.show()
```
- Mostra o gráfico que foi gerado com uma das funções plt;


````python
# EXEMPLO:
import matplotlib.pyplot as plt
estudantes = ['João', 'Maria', 'Pedro', 'Luisa']
notas = [8.5, 9, 6.5, 7]
plt.bar(estudantes, notas)
plt.show()
````

### Random

```
from random import choice
```
- Essa forma de importação consiste em importar apenas uma função especifica (choice) da biblioteca;
- `choice(lista)`: Escolhe um elemento de uma LISTA aleatoriamente;


````python
# EXEMPLO:
from random import choice
alunos = ['André', 'Lucas', 'Gabriela', 'Klelya']
aluno_alt = choice(alunos)
aluno_alt
````

````python
# EXEMPLO:
from random import choice
lista = [8, 12, 54, 23, 43, 1, 90, 87, 105, 77]
escolha = choice(lista)
print(escolha)
````

```
from random import choices
```
- Importa a função choices da biblioteca random;
- `choices(lista, k=n)`: Escolhe N elementos de uma LISTA aleatoriamente;


````python
# EXEMPLO:
from random import choices

frutas = ["maçã", "banana", "uva", "pêra",
          "manga", "coco", "melancia", "mamão",
          "laranja", "abacaxi", "kiwi", "ameixa"]

salada_choices = choices(frutas, k=3)

print(f'Sua salada vai ser composta de: {salada_choices[0]}, {salada_choices[1]} e {salada_choices[2]}')
````

```
from random import randrange
```
- `randrange(n)` ou `randrange(inicio, fim, salto)`:
    - Retorna um número aleatório de 0 a N;
    - Pode especificar o iinicio (incluso), fim (excluso) e o salto;



```
from random import sample
```
- `sample(lista, n)`: Retorna uma amostra de N elementos de uma LISTA;

````python
# EXEMPLO:
from random import randrange, sample
lista = []
for elemento in range(0, 20):
    lista.append(randrange(100))
amostra = sample(lista, 5)
print(amostra)
````

````python
# EXEMPLO:
from random import randrange
int_pst = randrange(100)
print(int_pst)
````

````python
# EXEMPLO:
nome = input('Insira seu nome: ')
token = randrange(1000, 10000, 2)
print(f'Olá, {nome}, o seu token de acesso é {token}! Seja bem-vindo(a)!')
````

````python
# EXEMPLO:
from random import sample

frutas = ["maçã", "banana", "uva", "pêra",
          "manga", "coco", "melancia", "mamão",

          "laranja", "abacaxi", "kiwi", "ameixa"]
salada_sample = sample(frutas, 3)

print(f'Sua salada vai ser composta de: {salada_sample[0]}, {salada_sample[1]} e {salada_sample[2]}')
````

```
from random import randint
```
- Importa a função randint da biblioteca random;
- `randint(a, b)`: Retorna um número inteiro aleatório entre a e b, incluindo os dois limites;

````python
# EXEMPLO:
from random import randint
participantes = int(input('Digite o número de participantes: '))
sorteio = randint(1, participantes)
print(f'O número sorteado foi {sorteio}')
````

### Math

```
from math import sqrt
```
- Importa a função sqrt da biblioteca math;
- `sqrt(n)`: Retorna a raíz quadrada de N;


````python
# EXEMPLO:
from math import sqrt
num = 16
print(sqrt(num))
````

````python
# EXEMPLO:
import math
numeros = [2, 8, 15, 23, 91, 112, 256]
for i in numeros:
    raiz = math.sqrt(i)
    if raiz // 1 == raiz:
        print(f'A raiz do numero {i} é de {raiz:.2f} e é inteira.')
    else:
        print(f'A raiz do numero {i} é de {raiz:.2f} e não é inteira.')
````

```
from math import ceil
```
- Importa a função ceil da biblioteca math;
- `ceil(n)`: Arredonda o valor de N para cima, gerando um número inteiro (sem parte decimal);

````python
# EXEMPLO:
from math import ceil
receita = 1000
unidade = 15
venda = receita / unidade
print(f'O valor da venda sem arredondamento: {venda}')
print(f'A empresa vendeu {ceil(venda)} unidades')
````

```
from math import pow
```
- Importa a função pow da biblioteca math;
- `pow(num1, num2)`: Retorna a potência do num1 elevado ao num2;

```
from math import pi
```
- Importa a função pi da biblioteca math;
- `pi`: Retorna o valor de PI como 3.141592653589793;

````python
# EXEMPLO:
from math import pow
num1 = int(input('Digite o primeiro valor: '))
num2 = int(input('Digite o segundo valor: '))
resultado = pow(num1, num2)
print(f'O resultado foi de: {resultado}')
````

````python
# EXEMPLO:
from math import pow, pi
raio = float(input('Insira o valor do raio (em metros) do jardim: '))
area = pi * pow(raio, 2)
valor = area * 25
print(f'O seu jardim tem {area:.2f}m² e custará R${valor:.2f}')
````

### Collections

```
from collections import Counter
```
- Importar a função Counter da biblioteca collections;
- `Counter(lista)`: Retorna um objeto que conta quantas vezes cada valor unico de uma lista aparece;


````python
# EXEMPLO:
from collections import Counter

estados = ['SP', 'ES', 'MG', 'MG', 'SP', 'MG', 'ES', 'ES', 'ES', 'SP', 'SP', 'MG', 'ES', 'SP', 'RJ', 'MG', 'RJ', 'SP', 'MG', 'SP', 'ES', 'SP', 'MG']

contador = dict(Counter(estados))
print(contador)
````
